```python
import itertools

bases = ['A','U','C','G']

k = 8

permutations = []

for p in itertools.product(bases, repeat=k):
    permutations.append(''.join(p))


import uuid

    
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import RNAAlphabet, IUPAC

all_TIS_seq = []
for string in permutations:
    all_TIS_seq.append(SeqRecord(Seq(string[0:6]+"AUG"+string[6:8],RNAAlphabet()),name = "",id= str(uuid.uuid4()), description = ""))
    
#     permutations[i][0:6] + "AUG"+ permutations[i][6:8]


```
